```python
top_5 = rf.groupby('country').sum()['Military expenditure (current USD)'].nlargest(5)
top_5 = list(top_5.index)
```

```python
top_5
```

```python
def plotter_1(top_5):
    plt.figure(figsize=(20, 8))
    plt.title('Military expenditure of 5 countries (max values) for different times', fontsize=18 )
    YEARS = [i for i in range(df['year'].min(), df['year'].max() + 1)]
    for top_country in top_5:
        country_data = df[df['country'] == top_country]
        sns.lineplot(x=country_data['year'], y=country_data['Military expenditure (current USD)'] , label=top_country)
    plt.legend(fontsize=18) 
    plt.xticks(rotation=30 , fontsize=18)
    plt.yticks(fontsize=18)
    plt.xlabel('year', fontsize=18)
    plt.ylabel('Military expenditure (current USD)' , fontsize=18)
    plt.grid()
plotter_1(top_5)
plt.show()
```

```python
tf = df.loc[df['Military expenditure (current USD)'] >0, :]
#Find the 5 countries with the smallest tota; military expenditure

smallest_5 = tf.groupby('country').sum()['Military expenditure (current USD)'].nsmallest(5)
smallest_5 = list(smallest_5.index)
def plotter_10(smallest_5):
    plt.figure(figsize=(20, 8))
    plt.title('5 countries with the lowest military spending for the period 1970-2020', fontsize=18)
    YEARS = [i for i in range(df['year'].min() , df['year'].max() + 1)]
    for country in smallest_5:
        country_data = df[df['country'] == country]
        sns.lineplot(x=country_data['year'], y=country_data['Military expenditure (current USD)'], label=country)
    plt.legend(fontsize=18)
    plt.xticks(rotation=30, fontsize=18)
    plt.yticks(fontsize=18)
    plt.xlabel('year', fontsize=18)
    plt.ylabel('Military expenditure (current USD )' , fontsize=18)
    plt.grid()
plotter_10(smallest_5)
plt.show()
```

```python
income_level_list = df['incomeLevel'].dropna().unique().tolist()
income_level_list.remove('Not classified')
income_level_list.remove('Aggregates')
income_level_list
```

```python
mf = pd.DataFrame(df.groupby('incomeLevel').sum()['Military expenditure (current USD)'])
mf.drop(index=['Aggregates', 'Not classified'], inplace=True)
explode = (0.05 , 0.05 , 0.05, 0.05)
plt.figure(figsize=(10 , 10))
plt.title('''The ratio of the tota; Military spending of countries with different levels of income of the proulation for the period 1970-2020''', fontsize =17)
plt.pie(mf['Military expenditure (current USD)'], labels=income_level_list, autopct='%1.1f%%', shadow= False,
       wedgeprops={'lw':1, 'ls': '--', 'edgecolor': 'k'}, rotatelabels=True, explode=explode,colors=sns.color_palette('pastel')[0:4],textprops={'fontsize':14})
plt.show()
```

## conclusion :
## The chart shows that low income countries have the highest total military spending

```python
df_update = df[df['country'].isin(['Low income' , 'Lower middle income' , 'upper middle income' , 'High income'])]
df_update = df_update.loc[:, ['country' , 'year' , 'Military expenditure (% of general government expenditure)']]

# Function to plot military expenditure for a given income level
def plotter_3(income_level):
    income_data = df_update[df_update['country'] == income_level]
    sns.lineplot(x=income_data['year'], y=income_data['Military expenditure (% of general government expenditure)'],
                 label=f'{income_level} level')

# Set up the plot
plt.figure(figsize=(20, 10))
plt.title('Military expenditure (% of general government expenditure) for countries with different income levels', fontsize=17)
plt.xlabel('Year', fontsize=17)
plt.ylabel('Military expenditure (% of general government expenditure)', fontsize=17)
plt.grid()

# Plot for each income level
plotter_3('High income')
plotter_3('Upper middle income')
plotter_3('Lower middle income')
plotter_3('Low income')

# Customize the ticks and legend
plt.xticks(fontsize=17)
plt.yticks(fontsize=17)
plt.legend(fontsize=17)

# Show the plot
plt.show()
```

## conclusion :
## And low income level countires lead the way , although military expenditure (% of general government expenditure ) has declined over the past 15 years
